<h1 align="center">AgenticAuthor</h1>
<h2 align="center">Autonomous Research & Article Writing with Open Sourced LLMs and CrewAI</h1>

### 1. Define Agents

```python
from crewai import Agent, Task, Crew, Process
from langchain_ollama import ChatOllama

researcher = Agent(
    role='Senior Researcher',
    goal='Uncover groundbreaking insights on a given topic',
    backstory='A seasoned analyst with a knack for finding hidden gems in data.',
    verbose=True,
    allow_delegation=False,
    llm = ChatOllama(model = 'ollama/gemma3:12b', base_url="http://localhost:11434")
)

writer = Agent(
    role='Content Writer',
    goal='Craft compelling and informative articles',
    backstory='A creative wordsmith known for engaging storytelling.',
    verbose=True,
    allow_delegation=True,
    llm = ChatOllama(model = 'ollama/gpt-oss:20b', base_url="http://localhost:11434")
)
```

### 2. Define Tasks

```python
research_task = Task(
    description='Conduct in-depth research on the latest AI trends.',
    agent=researcher,
    expected_output='A detailed report outlining key AI trends and their potential impact.'
)

writing_task = Task(
    description=
    """
        Write a detailed blog post based on the research report. Respond with only the blog post content.

        Do not include any prefaces, explanations, acknowledgments, conclusions about the writing itself, or meta-commentary (e.g., â€œHere is the blog post,â€ â€œIn this article,â€ â€œI will explainâ€).
        
        Do not reference yourself, the prompt, or the writing process.
        
        Do not add titles like â€œBlog Post:â€ or similar meta labels.
        
        Write in natural, reader-ready prose suitable for direct publication.
    """,
    agent=writer,
    expected_output="""A well-structured and engaging blog post about AI trends with heading, subheadings and tables as needed.""",
    markdown_text = True
)
```

### 3. Assemble the Crew

```python
my_crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, writing_task],
    process=Process.sequential,
    verbose=True
)
```

### 4. Kick off the Crew's work

```python
result = my_crew.kickoff()
```

### 5. Generated Article

```python
from rich.console import Console
from rich.markdown import Markdown

console = Console()
md = Markdown(result.raw)
console.print(md)
```
